#Using nltk tnt tagger

```python
from nltk.tag import tnt
from nltk.corpus import indian
import nltk
import pandas as pd
import numpy as np
nltk.download('punkt') 
nltk.download('indian')
```

```python

text = "इराक के विदेश मंत्री ने अमरीका के उस प्रस्ताव का मजाक उड़ाया है , जिसमें अमरीका ने संयुक्त राष्ट्र के प्रतिबंधों को इराकी नागरिकों के लिए कम हानिकारक बनाने के लिए कहा है ।" #"मेरा नाम वैभव है"
```

```python
def hindi_model():
    train_data = indian.tagged_sents('hindi.pos')
    tnt_pos_tagger = tnt.TnT()
    tnt_pos_tagger.train(train_data)
    return tnt_pos_tagger
```

```python
print(nltk.word_tokenize(text))
```

```python
model = hindi_model()
new_tagged = (model.tag(nltk.word_tokenize(text)))
#print(new_tagged)
array=np.array(new_tagged)
for i in array:
    print(i[0]," ==> ", i[1]," ")
```

```python
with open("output.txt","a") as output_file:
    output_file.write("\n[INPUT]\n")
    output_file.write(text)
    output_file.write("\n[OUTPUT]\n")
    output_file.write(str(new_tagged))
```

```python
import codecs
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tag import tnt
from nltk.corpus import indian

# Read the text file
with codecs.open('test_data.txt', mode='r', encoding='utf-8') as f:
    data = f.read()

# Tokenize the sentences
sentences = sent_tokenize(data)

# Create a TNT POS tagger for Hindi
tnt_pos_tagger = tnt.TnT()

# Train the POS tagger using the Indian Corpus in NLTK
train_data = indian.tagged_sents('hindi.pos')
tnt_pos_tagger.train(train_data)

# Perform POS tagging on each sentence
for sentence in sentences:
    # Tokenize the words in the sentence
    words = word_tokenize(sentence)
    # Perform POS tagging on the words in the sentence
    tagged_words = tnt_pos_tagger.tag(words)
    # Print the tagged words
    print(tagged_words)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tag import tnt
from nltk.corpus import indian


# Load the POS tagger for Hindi
nltk.download('indian')

# Load the data from the text file
with open('test_data.txt', 'r', encoding='utf-8') as file:
    data = file.read()

# Tokenize the sentences
sentences = sent_tokenize(data, 'hi')

# Create a TNT POS tagger for Hindi
tagger = tnt.TnT()

# Train the tagger on the Hindi POS tagged dataset
nltk.download('hindi.pos')

with open('hindi.pos', 'r', encoding='utf-8') as pos_file:
    tagged_words = [line.split() for line in pos_file.readlines()]

tagger.train(tagged_words)

# Tag each word in each sentence and write to a new file
with open('pos_tagged_hindi.txt', 'w', encoding='utf-8') as outfile:
    for sentence in sentences:
        words = word_tokenize(sentence, 'hindi')
        tagged_words = tagger.tag(words)
        outfile.write(' '.join([f"{word}/{tag}" for word, tag in tagged_words]) + '\n')
```
